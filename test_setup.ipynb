```python
#@title 🎙️ Baseline Test: English Voice Design
import torch
import soundfile as sf
from qwen_tts.inference.qwen3_tts_model import Qwen3TTSModel

# 1. Load the Model
# We remove 'flash_attention_2' since we are using the standard install
model = Qwen3TTSModel.from_pretrained(
    "Qwen/Qwen3-TTS-12Hz-1.7B-VoiceDesign",
    device_map="cuda:0",
    dtype=torch.bfloat16
)

# 2. English Inference
# We change the text and the 'instruct' to English
text_to_generate = "Welcome to Project Vaders. We are currently calibrating the deep learning speaker encoder for the CS 7643 research project."

wavs, sr = model.generate_voice_design(
    text=text_to_generate,
    language="English",
    instruct="A happy person talking about a professional success"
)

# 3. Save the Result
output_filename = "vaders_baseline_english.wav"
sf.write(output_filename, wavs[0], sr)

print(f"✅ Success! Audio saved as {output_filename}")
```
